```python
from sklearn.pipeline import Pipeline                                             #Model Comparison using Pipelines
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, OneHotEncoder

X_train, X_test, y_train, y_test = train_test_split(x_cleaned, y_cleaned, test_size=0.2, random_state=42)

models = {
    "LogisticRegression": LogisticRegression(),
    "RandomForest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "GradientBoosting": GradientBoostingClassifier()
}

results = {}

for name, model in models.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"{name} Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt

plt.bar(results.keys(), results.values(), color='skyblue')
plt.ylabel('Accuracy Score')
plt.title('Model Comparsion')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib

X_train, X_test, y_train, y_test = train_test_split(x_cleaned, y_cleaned, test_size=0.2, random_state=42)


models = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "RandomForest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "GradientBoosting": GradientBoostingClassifier()
}

results = {}


for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    results[name] = acc
    print(f"{name}: {acc:.4f}")


best_model_name = max(results, key=results.get)
best_model = models[best_model_name]
print(f"\n✅ Best model: {best_model_name} with accuracy {results[best_model_name]:.4f}")


joblib.dump(best_model, "best_model.pkl")
print("✅ Saved best model as best_model.pkl")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline


try:

    full_data_for_preprocessing = pd.read_csv("adult 3.csv")

    full_data_for_preprocessing.occupation.replace({'?':'Others'},inplace=True)
    full_data_for_preprocessing.workclass.replace({'?':'Not listed'},inplace=True)
    full_data_for_preprocessing=full_data_for_preprocessing[full_data_for_preprocessing['workclass']!= 'Without-pay']
    full_data_for_preprocessing=full_data_for_preprocessing[full_data_for_preprocessing['workclass']!= 'Never-worked']
    full_data_for_preprocessing=full_data_for_preprocessing[full_data_for_preprocessing['education']!= '1st-4th']
    full_data_for_preprocessing=full_data_for_preprocessing[full_data_for_preprocessing['education']!= 'Preschool']
    full_data_for_preprocessing=full_data_for_preprocessing[full_data_for_preprocessing['education']!= '5th-6th']
    full_data_for_preprocessing = full_data_for_preprocessing[(full_data_for_preprocessing['age'] <= 75) & (full_data_for_preprocessing['age'] >= 17)].copy()
    full_data_for_preprocessing.drop(columns=['education'],inplace=True)

    X_full = full_data_for_preprocessing.drop(columns=['income'])
    y_full = full_data_for_preprocessing['income']


    label_encoders = {}
    categorical_cols = ['workclass', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country']
    for col in categorical_cols:
        le = LabelEncoder()
        X_full[col] = le.fit_transform(X_full[col])
        label_encoders[col] = le


    scaler = MinMaxScaler()
    numerical_cols = ['age', 'fnlwgt', 'educational-num', 'capital-gain', 'capital-loss', 'hours-per-week']
    X_scaled_full = X_full.copy()
    X_scaled_full[numerical_cols] = scaler.fit_transform(X_full[numerical_cols])



    X_train, X_test, y_train, y_test = train_test_split(X_scaled_full, y_full, test_size=0.2, random_state=42, stratify=y_full)

except Exception as e:
    st.error(f"Error during initial data loading or preprocessing: {e}. Please ensure 'adult 3.csv' is in the correct directory.")
    st.stop()


try:
    model = joblib.load("best_model.pkl")
except Exception as e:
    st.error(f"Error loading the best_model.pkl: {e}. Make sure the model file exists and is accessible.")
    st.stop()


st.set_page_config(page_title="Employee Salary Classification", page_icon="💼", layout="wide")

st.title("💼 Employee Salary Predictor")
st.markdown("Predict whether an employee earns >50K or ≤50K based on input features.")


tab1, tab2, tab3 = st.tabs(["Data Exploration", "Single Prediction", "Batch Prediction & Model Info"])


with tab1:
    st.header("📊 Data Exploration")
    st.markdown("Here you can see various visualizations of the dataset.")


    if 'full_data_for_preprocessing' in locals():

        st.subheader("Distribution of Income")
        fig, ax = plt.subplots(figsize=(6, 4))
        sns.countplot(data=full_data_for_preprocessing, x='income', ax=ax)
        ax.set_title('Distribution of Income')
        ax.set_xlabel('Income Class')
        ax.set_ylabel('Count')
        st.pyplot(fig)


        st.subheader("Age Distribution by Income")
        fig, ax = plt.subplots(figsize=(10, 6))
        sns.histplot(data=full_data_for_preprocessing, x='age', hue='income', kde=True, palette='viridis', ax=ax)
        ax.set_title('Age Distribution by Income')
        ax.set_xlabel('Age')
        ax.set_ylabel('Count')
        st.pyplot(fig)


        st.subheader("Hours per Week by Income")
        fig, ax = plt.subplots(figsize=(8, 5))
        sns.boxplot(data=full_data_for_preprocessing, x='income', y='hours-per-week', palette='plasma', ax=ax)
        ax.set_title('Hours per Week by Income')
        ax.set_xlabel('Income Class')
        ax.set_ylabel('Hours per Week')
        st.pyplot(fig)


        if 'education' in pd.read_csv("adult 3.csv").columns:
            original_data_for_edu = pd.read_csv("adult 3.csv")

            original_data_for_edu=original_data_for_edu[original_data_for_edu['education']!= '1st-4th']
            original_data_for_edu=original_data_for_edu[original_data_for_edu['education']!= 'Preschool']
            original_data_for_edu=original_data_for_edu[original_data_for_edu['education']!= '5th-6th']

            st.subheader("Education Level vs. Income")
            fig, ax = plt.subplots(figsize=(12, 7))
            sns.countplot(data=original_data_for_edu, x='education', hue='income', palette='coolwarm',
                          order=original_data_for_edu['education'].value_counts().index, ax=ax)
            ax.set_title('Education Level vs. Income')
            ax.set_xlabel('Education Level')
            ax.set_ylabel('Count')
            ax.tick_params(axis='x', rotation=45)
            plt.tight_layout()
            st.pyplot(fig)
        else:
            st.info("Education level visualization skipped as 'education' column was removed during preprocessing.")


        st.subheader("Income Proportion by Occupation")

        occupation_income = full_data_for_preprocessing.groupby('occupation')['income'].value_counts(normalize=True).unstack()
        fig, ax = plt.subplots(figsize=(14, 7))
        occupation_income.plot(kind='bar', stacked=True, cmap='viridis', ax=ax)
        ax.set_title('Income Proportion by Occupation')
        ax.set_xlabel('Occupation')
        ax.set_ylabel('Proportion')
        ax.tick_params(axis='x', rotation=90)
        plt.tight_layout()
        st.pyplot(fig)


        st.subheader("Correlation Matrix of Features (after encoding & scaling)")

        X_y_combined = X_scaled_full.copy()
        X_y_combined['income'] = y_full.map({'<=50K': 0, '>50K': 1})

        fig, ax = plt.subplots(figsize=(12, 10))
        sns.heatmap(X_y_combined.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f", ax=ax)
        ax.set_title('Correlation Matrix of Features')
        st.pyplot(fig)

    else:
        st.warning("Data for visualization could not be loaded. Please check data file.")


with tab2:
    st.header("🔮 Make a Single Prediction")
    st.markdown("Enter details to predict the salary class for an individual employee.")

    st.sidebar.header("Input Employee Details")


    age = st.sidebar.slider("Age", 17, 75, 30)
    workclass = st.sidebar.selectbox("Workclass", full_data_for_preprocessing['workclass'].unique().tolist())
    fnlwgt = st.sidebar.number_input("Fnlwgt (Final Weight)", min_value=int(X_full['fnlwgt'].min()), max_value=int(X_full['fnlwgt'].max()), value=200000)
    educational_num = st.sidebar.slider("Educational Num", int(X_full['educational-num'].min()), int(X_full['educational-num'].max()), 9)
    marital_status = st.sidebar.selectbox("Marital Status", full_data_for_preprocessing['marital-status'].unique().tolist())
    occupation = st.sidebar.selectbox("Occupation", full_data_for_preprocessing['occupation'].unique().tolist())
    relationship = st.sidebar.selectbox("Relationship", full_data_for_preprocessing['relationship'].unique().tolist())
    race = st.sidebar.selectbox("Race", full_data_for_preprocessing['race'].unique().tolist())
    gender = st.sidebar.selectbox("Gender", full_data_for_preprocessing['gender'].unique().tolist())
    capital_gain = st.sidebar.number_input("Capital Gain", min_value=int(X_full['capital-gain'].min()), max_value=int(X_full['capital-gain'].max()), value=0)
    capital_loss = st.sidebar.number_input("Capital Loss", min_value=int(X_full['capital-loss'].min()), max_value=int(X_full['capital-loss'].max()), value=0)
    hours_per_week = st.sidebar.slider("Hours per week", int(X_full['hours-per-week'].min()), int(X_full['hours-per-week'].max()), 40)
    native_country = st.sidebar.selectbox("Native Country", full_data_for_preprocessing['native-country'].unique().tolist())


    input_data_single = pd.DataFrame([{
        'age': age,
        'workclass': workclass,
        'fnlwgt': fnlwgt,
        'educational-num': educational_num,
        'marital-status': marital_status,
        'occupation': occupation,
        'relationship': relationship,
        'race': race,
        'gender': gender,
        'capital-gain': capital_gain,
        'capital-loss': capital_loss,
        'hours-per-week': hours_per_week,
        'native-country': native_country
    }])

    st.write("### 🔎 Input Data for Prediction")
    st.write(input_data_single)

    if st.button("Predict Salary Class"):
        processed_input_single = input_data_single.copy()


        for col in categorical_cols:
            if col in processed_input_single.columns:

                try:
                    processed_input_single[col] = label_encoders[col].transform(processed_input_single[col])
                except ValueError:
                    st.error(f"Unseen category in '{col}'. Please select a valid option.")
                    st.stop()



        if numerical_cols:
            processed_input_single[numerical_cols] = scaler.transform(processed_input_single[numerical_cols])

        try:
            prediction = model.predict(processed_input_single)
            st.success(f"✅ Predicted Salary Class: **{prediction[0]}**")
        except Exception as e:
            st.error(f"Prediction error: {e}. Ensure input data is correctly preprocessed and matches model expectations.")



with tab3:
    st.header("📦 Batch Prediction & Model Information")
    st.markdown("Upload a CSV file to get predictions for multiple employees, and view information about the trained models.")

    st.markdown("#### 📂 Batch Prediction")
    uploaded_file = st.file_uploader("Upload a CSV file for batch prediction", type="csv")

    if uploaded_file is not None:
        batch_data = pd.read_csv(uploaded_file)
        st.write("Uploaded data preview:", batch_data.head())


        try:
            processed_batch_data = batch_data.copy()


            processed_batch_data.occupation.replace({'?':'Others'},inplace=True)
            processed_batch_data.workclass.replace({'?':'Not listed'},inplace=True)


            if 'education' in processed_batch_data.columns:
                processed_batch_data.drop(columns=['education'], inplace=True)


            for col in categorical_cols:
                if col in processed_batch_data.columns:

                    processed_batch_data[col] = processed_batch_data[col].apply(
                        lambda x: label_encoders[col].transform([x])[0] if x in label_encoders[col].classes_ else -1
                    )



            if numerical_cols:
                 if all(col in processed_batch_data.columns for col in numerical_cols):
                    processed_batch_data[numerical_cols] = scaler.transform(processed_batch_data[numerical_cols])
                 else:
                    st.warning("Numerical columns missing in batch data. Scaling skipped.")


            batch_preds = model.predict(processed_batch_data)
            batch_data['PredictedClass'] = batch_preds
            st.write("✅ Predictions (first 5 rows):")
            st.dataframe(batch_data.head())

            csv = batch_data.to_csv(index=False).encode('utf-8')
            st.download_button("Download Predictions CSV", csv, file_name='predicted_classes.csv', mime='text/csv')

        except Exception as e:
            st.error(f"Error processing batch data: {e}. Please ensure your CSV columns match the expected input for the model and that the data is clean.")


    st.markdown("---")
    st.markdown("#### ⚙️ Model Performance & Details")
    st.write("Here are the accuracy scores and classification reports for various models trained on the dataset.")


    if 'X_train' in locals() and 'X_test' in locals():
        models_for_report = {
            "LogisticRegression": LogisticRegression(max_iter=1000),
            "RandomForest": RandomForestClassifier(random_state=42),
            "KNN": KNeighborsClassifier(),
            "SVM": SVC(random_state=42),
            "GradientBoosting": GradientBoostingClassifier(random_state=42)
        }

        for name, model_report in models_for_report.items():
            model_report.fit(X_train, y_train)
            y_pred_report = model_report.predict(X_test)
            acc_report = accuracy_score(y_test, y_pred_report)
            st.subheader(f"{name} Performance:")
            st.write(f"Accuracy: {acc_report:.4f}")
            st.text("Classification Report:")
            st.code(classification_report(y_test, y_pred_report))
    else:
        st.warning("Model performance reports cannot be generated as training data was not prepared.")
```
